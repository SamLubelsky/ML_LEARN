```python
import os
import tarfile
import urllib.request
from pathlib import Path
BASE_PATH = Path("C:/Users/sammy/ML_LEARN")
FULL_PATH = BASE_PATH / "datasets" / "spam"
def get_data(filename):
    PATH = FULL_PATH / filename
    tar = tarfile.open(PATH)
    tar.extractall(FULL_PATH)
get_data("20021010_easy_ham.tar.bz2")
get_data("20021010_spam.tar.bz2")
get_data("20030228_spam_2.tar.bz2")
```

```python
spam_files = os.listdir(FULL_PATH / "spam")
spam_2_files = os.listdir(FULL_PATH / "spam_2")
spam = np.empty(len(spam_files) + len(spam_2_files), dtype=object)
i = 0
for file in spam_files:
    PATH = FULL_PATH / "spam" / file
    with open(PATH, 'r', encoding = 'Latin-1') as f:
        spam[i] = f.read()
    i += 1
for file in spam_2_files:
    PATH = FULL_PATH / "spam_2" / file
    with open(PATH, 'r', encoding = 'Latin-1') as f:
        spam[i] = f.read()
    i += 1
ham_files = os.listdir(FULL_PATH / "easy_ham")
ham = np.empty(len(ham_files), dtype=object)
i = 0
for file in ham_files:
    PATH = FULL_PATH / "easy_ham" / file
    with open(PATH, 'r', encoding = 'Latin-1') as f:
        ham[i] = f.read()
    i += 1
```

```python
#data cleaning
cleaned_spam = np.empty(len(spam), dtype = object)
i = 1
for mail in spam:
    end_location = mail.find('\n\n')
    cleaned_mail = mail[end_location + 2:]
    cleaned_spam[i] = cleaned_mail
    i += 1
cleaned_ham = np.empty(len(ham), dtype = object)
i = 0
for mail in ham:
    end_location = mail.find('\n\n')
    cleaned_mail = mail[end_location + 2:]
    cleaned_ham[i] = cleaned_mail
    i += 1
```

```python
from sklearn.model_selection import train_test_split
TEST_SIZE = 0.2
X = np.append(cleaned_spam,cleaned_ham)
y = np.append(np.full(len(cleaned_spam), True), np.full(len(cleaned_ham), False))
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = TEST_SIZE, random_state = 42)
```

```python
def is_valid(word):
    return word.isalpha() and len(word) < 10
```

```python
def clean(word):
    word = word.lower()
    word = word.replace("'", "")
    return word
```

```python
X_train_prepared = lil_matrix((len(X_train), len(words_dict)))
i = 0
for mail in X_train:
    words = mail.split()
    for word in words:
        word = clean(word)
        if(is_valid(word)):
            word_pos = words_dict[word]
            X_train_prepared[i, word_pos] = 1
    i += 1  
```

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state = 42)
sgd_clf.fit(X_train_prepared, y_train)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, X_train_prepared, y_train, cv = 3, scoring = 'accuracy')
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_predict
y_train_pred = cross_val_predict(sgd_clf, X_train_prepared, y_train, cv = 3)
confusion_matrix(y_train, y_train_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_train, y_train_pred)
```
